# Data Collection

### Getting the initial postal codes and lat/long coordinates for Toronto

Two sources of data were used here. The first is a wikipedia table of dataframes in [Toronto](https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M). The second was a provided dataset of latitude and longitude [coordinates](http://cocl.us/Geospatial_data) for the postal codes

```python
#wikipedia link for postal codes for toronto and reading table into dataframe
wiki_link = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
canada = pd.read_html(wiki_link, header = 0)[0]
canada = canada[canada['Borough'] != 'Not assigned']
canada['Neighbourhood'] = np.where(canada['Neighbourhood'] == 'Not assigned', canada['Borough'], canada['Neighbourhood'])


#merging the two data frames created
ca_n = canada.groupby(['Postcode'])['Neighbourhood'].apply(','.join)
ca_n = pd.DataFrame(ca_n, columns = ['Neighbourhood']).reset_index()

ca_b = canada[['Postcode','Borough']].drop_duplicates(inplace = False)
canada = ca_b.join(ca_n.set_index('Postcode'), on = 'Postcode').reset_index(drop = True)
#canada is dataframe with postal code, neighbourhoods, and borough

#source had dataset with latitude and longitude for canadian postal codes
#adding latitude and longitude to canada table
canada_post = pd.read_csv("http://cocl.us/Geospatial_data")
canada_post = canada.join(canada_post.set_index('Postal Code'), on = 'Postcode').reset_index(drop=True)

#getting number of neighbourhoods for postal code
canada_post['num_N'] = canada_post['Neighbourhood'].apply(lambda h: len(h.split(',')))
print(canada_post.head())
canada.shape
```


Setting up functions for getting restaurants of certain categories using Foursquare api 

```python
#functions for getting venues near each location
import requests



#gets category of restaurant and retruns two booleans. Is restaurant, and if the rest is a specific category
def is_restaurant(categories, specific_filter=None):
    restaurant_words = ['restaurant', 'diner', 'taverna', 'steakhouse']
    restaurant = False
    specific = False
    for c in categories:
        category_name = c[0].lower()
        category_id = c[1]
        for r in restaurant_words:
            if r in category_name:
                restaurant = True
        if 'fast food' in category_name:
            restaurant = False
        if not(specific_filter is None) and (category_id in specific_filter):
            specific = True
            restaurant = True
    return restaurant, specific

def get_categories(categories):
    return [(cat['name'], cat['id']) for cat in categories]


#getting rid of references to the city, province, and country
def format_address(location):
    address = ', '.join(location['formattedAddress'])
    address = address.replace(', Toronto', '')
    address = address.replace(', Canada', '')
    address = address.replace(', Ontario', '')
    address = address.replace(', CA', '')
    return address


#does the Fourspace call with query for locations near latitude and longitude
def get_venues_near_location(lat, lon, category, client_id, client_secret, radius=500, limit=100):
    version = '20180724'
    url = 'https://api.foursquare.com/v2/venues/explore?client_id={}&client_secret={}&v={}&ll={},{}&categoryId={}&radius={}&limit={}'.format(
        client_id, client_secret, version, lat, lon, category, radius, limit)
    try:
        results = requests.get(url).json()['response']['groups'][0]['items']
        venues = [(item['venue']['id'],
                   item['venue']['name'],
                   get_categories(item['venue']['categories']),
                   (item['venue']['location']['lat'], item['venue']['location']['lng']),
                   format_address(item['venue']['location']),
                   item['venue']['location']['distance']) for item in results]
    except:
        venues = []
    return venues
```

```python
#finally function to convert Fourspace venue data into data with just specific category I want
def get_restaurants(lats, lons):
    
    #These two variables will be left empty on github for privacy reasons. Thus function will fail unless you put 
    #your own client id and secret
    foursquare_client_id = 'YTNGUBZEABCJMMAR45HTNGQOOSDXKTNTSXIRCUTZNEZYV4CB'
    foursquare_client_secret = 'XOIYQSAF4OAAUFOZXXQPY4PAVNQXF1D1ZWUDINJQSZUP5HMT'
    
    food_category = '4d4b7105d754a06374d81259'
    restaurant_cat = ['4bf58dd8d48988d142941735','56aa371be4b08b9a8d573568','52e81612bcbc57f1066b7a03','4bf58dd8d48988d145941735',
                         '4eb1bd1c3b7b55596b4a748f','52e81612bcbc57f1066b79fb','52af0bd33cf9994f4e043bdd','4deefc054765f83613cdba6f',
                         '4bf58dd8d48988d111941735','4bf58dd8d48988d113941735','4bf58dd8d48988d156941735','4eb1d5724b900d56c88a45fe',
                         '4bf58dd8d48988d1d1941735','56aa371be4b08b9a8d57350e','4bf58dd8d48988d149941735','52af39fb3cf9994f4e043be9',
                         '4bf58dd8d48988d14a941735']
    
    restaurants = {}
    cat_restaurants = {}
    location_restaurants = []
    
    for lat, lon in zip(lats, lons):
        # Using radius=500 with dictionaries to remove any duplicates resulting from area overlaps
        venues = get_venues_near_location(lat, lon, food_category, foursquare_client_id, foursquare_client_secret, radius=500, limit=100)
        area_restaurants = []
        for venue in venues:
            venue_id = venue[0]
            venue_name = venue[1]
            venue_categories = venue[2]
            venue_latlon = venue[3]
            venue_address = venue[4]
            venue_distance = venue[5]
            is_res, is_cat = is_restaurant(venue_categories, specific_filter= restaurant_cat)
            if is_res:
                restaurant = (venue_id, venue_name, venue_latlon[0], venue_latlon[1], venue_address, venue_distance, is_cat)
                if venue_distance<=500:
                    area_restaurants.append(restaurant)
                restaurants[venue_id] = restaurant
                if is_cat:
                    cat_restaurants[venue_id] = restaurant
        location_restaurants.append(area_restaurants)
    return restaurants, cat_restaurants, location_restaurants
```

## Building the Linear Models

Using the standard library sklearn to import a linear regression model. For the purposes of this exercise, checking the assumptions of the model will be ignored. Also, there will be no splitting in training/test/validation. The objective of the two linear models will be to train over whole dataset, then predict over that same dataset (not new data set). 

The difference between the predicted value and the actual value will be used to go to the next stage of the analysis:
1. If the predicted value = actual value, then that postcode will be assumed to be saturated
2. If the predicted value < actual value, then that postcode will be thrown out of the pool of possible locations
3. If the predicted value > actual value, then that postcode will be assumed to be a possible target

This will be done with both linear models:
1. If both are considered overpredicted, then that postcode will be analyzed for possible locations
2. If only number of restaurants is overpredicted, then the postcode will be ignored
3. If only number of asian restaurants is overpredicted, then the postcodes will be ignored

```python
#adding downtown variable
canada_post['Downtown'] = np.where(canada_post['Borough'] == 'Downtown Toronto',1,0)
canada_post.head()
```

```python
#Checking a seaborn plot for downtown variable

fig, axs = plt.subplots(1,2,figsize=(10, 5))
axs[0].title.set_text('Downtown vs. not Downtown for restaurant #')
sns.boxplot(x = 'Downtown', y = 'Num_rest', data = canada_post, ax = axs[0])
axs[1].title.set_text('Downtown vs. not Downtown for asian #')
sns.boxplot(x = 'Downtown', y = 'Asian_count', data = canada_post, ax = axs[1])
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

#X = canada_post[['Latitude','Longitude','Pop_2016','Homes_occupied','num_N', 'Downtown']]
X = canada_post[['Latitude','Longitude','Pop_2016','Homes_occupied','num_N']].copy()
X['num_N_2'] = X['num_N']*X['num_N']
X = scaler.fit_transform(X)
X = pd.DataFrame(data = X, columns = ['Latitude','Longitude','Pop_2016','Homes_occupied','num_N','num_N_2'])
X['Downtown'] = canada_post['Downtown']
Y_asian = canada_post['Asian_count']
Y_rest = canada_post['Num_rest']
```

```python
X.head()
```

```python
from sklearn.linear_model import LinearRegression

lm_general = LinearRegression().fit(X,Y_rest)
lm_asian = LinearRegression().fit(X,Y_asian)
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error

Y_pred_g = lm_general.predict(X)
Y_pred_a = lm_asian.predict(X)

print('The R2 score for general model is:', r2_score(Y_rest, Y_pred_g))
print('The MSE for general model is:', mean_squared_error(Y_rest, Y_pred_g))

print('\n')

print('The R2 score for asian model is:', r2_score(Y_asian, Y_pred_a))
print('The MSE for asian model is:', mean_squared_error(Y_asian, Y_pred_a))
```

```python
new_data = canada_post.copy()

new_data['rest_pred'] = Y_pred_g - Y_rest
new_data['asian_pred'] = Y_pred_a - Y_asian

#greater than one demands that there should be a difference in predicted/actual of one full restaurant
#After looking with 1, a small set of the data had >6 difference, filtering to just those postcodes
new_data['f_crit'] = np.where(new_data['rest_pred'] > 5, True, False)
new_data['s_crit'] = np.where(new_data['asian_pred'] > 1, True, False)
```

```python
first_choice = new_data.query('f_crit == True & s_crit == True & Num_rest > 0')
first_choice.reset_index(inplace = True)
```

Now that I have the final set of restaurants. I will use Kmeans to cluster the addresses where these restaurants are congregated. The basis is that these are areas that, based on the linear model, should have more restaurants. Since people go to common areas to get food (around break hours and for dinner), these will be the locations they will go to. This does have a trade-off. Higher traffic in these locations vs. higher competition. 

```python
from sklearn.cluster import KMeans

clusterer = KMeans(random_state = 42, n_clusters = 11)

clusterer.fit(final_set[['Latitude','Longitude']])
```

```python
#getting the cluster centers on a map
To_coord = [43.6832, -79.3832]

themap = folium.Map(location = To_coord, zoom_start = 12)

for i in range(10):
    #this places markers at the cluster centers
    folium.CircleMarker(clusterer.cluster_centers_[i],radius = 3,popup=clusterer.cluster_centers_[i],
                       color = 'green', fill = True, fill_color = 'green', fill_opacity = 1).add_to(themap)



for i in range(len(final_set)):
    #this places smaller markers for an asian restaurant near the cluster
    if(final_set.iloc[i]['asian'] == True):
        folium.CircleMarker([final_set.iloc[i]['Latitude'], final_set.iloc[i]['Longitude']],radius = 1,
                            color = 'red', fill = True, fill_color = 'red', fill_opacity = 1).add_to(themap)

themap
```

2. Hart House Circle, [Bloor Street Culture Corridor](https://www.seetorontonow.com/listings/bloor-st-culture-corridor/), Toronto, M5S 3H3. This is part of a diverse cultural district with many distinct, world-class, cultural and art organizations. That it has limited number of asian restaurants means this is a space to open a new cultural theme in a district designed to promote multiculturalism. It is also nearby an university and is downtown which means it will be a highly trafficked place.

3. Ashby Place, [Corktown](https://en.wikipedia.org/wiki/Corktown,_Toronto), Toronto,M51 1G1.This is an old residential district close to the heart of downtown Toronto. It is currently in the process of being gentrified, with many of its residential buildings being torn down in high value bidding for the land. There are a number of vacant industrial buildings there that have been repurposed for restaurants and other shops. That it is now being gentrified and converted into a commercial location makes it great for staking out an initial location for a market. The downside of this location is that there will be high competition to grab space as it is renovating.

4. Kingston Road, [Cliffside](http://www.torontoneighbourhoods.net/neighbourhoods/scarborough/cliffside?overview), Toronto, M1N 4A7. This is an upscale suburban housing area with a number of parks and waterfront areas. Much of the population are of descent from the British isles, which might provide a desire for exotic/ethnic food. Its distance from the highly trafficked areas of downtown Toronto does mean less demand than the other areas, but it does also have potential competition.
